```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
#Read the data
cars = pd.read_csv("ToyotaCorolla.csv")
cars.head()
```

```python
cars1=cars.drop(cars.columns.difference(["Price","Age_08_04","KM","HP","cc","Quarterly_Tax"]), axis=1)
```

```python
cars1.head()
```

```python
cars1.info()
```

```python
cars1.corr()
```

# Scatterplot between variables along with histograms

```python
#Format the plot background and scatter plots for all the variables
sns.set_style(style='darkgrid')
sns.pairplot(cars1)
```

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Price~Age_08_04+KM+HP+cc+Quarterly_Tax',data=cars1).fit()
```

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj,model.aic)
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(cars1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```

```python
car2=cars1.drop(cars1.index[[80]],axis=0)
```

```python
car3=car2.reset_index()
```

```python
car4=car3.drop(['index'],axis=1)
```

```python
import numpy as np
final_ml_V = smf.ols('Price~np.log(Age_08_04)+KM+HP+cc+Quarterly_Tax',data=car4).fit()
```

```python
model_influence_V = final_ml_V.get_influence()
(c_V, _) = model_influence_V.cooks_distance
```

```python
fig= plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(car4)),np.round(c_V,3));
plt.xlabel('Row index')
plt.ylabel('Cooks Distance');
```

```python
(np.argmax(c_V),np.max(c_V))
```
